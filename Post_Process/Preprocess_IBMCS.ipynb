```python
import pandas as pd
```

```python
df = pd.read_csv("../Dataset/IBM_Debater_(R)_CS_EACL-2017.v1/claim_stance_dataset_v1.csv")
```

```python
df
```

```python
col = ["claims.stance","x","y", "topicText", "claims.claimCorrectedText", "split"]
df = pd.DataFrame(df,columns=col)
```

```python
df = df.fillna('0')
```

```python
df["claims.stance"][df["claims.stance"]=="PRO"] = 1
df["claims.stance"][df["claims.stance"]=="CON"] = 0
```

```python
df
```

```python
test = df[df["split"]=="test"]
train = df[df["split"]=="train"]
```

```python
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
```

```python
train.rename(columns={'claims.stance':'Label','topicText':'claim', 'claims.claimCorrectedText':'perspective'},inplace=True) 
```

```python
train = train.drop("split",axis=1)
```

```python
test.rename(columns={'claims.stance':'Label','topicText':'claim', 'claims.claimCorrectedText':'perspective'},inplace=True) 
```

```python
test = test.drop("split",axis=1)
```

```python
test
```

```python
train.to_csv("./Dataset/ibmcs/train.tsv", sep='\t',index=False)
```

```python
test.to_csv("./Dataset/ibmcs/test.tsv", sep='\t',index=False)
```
